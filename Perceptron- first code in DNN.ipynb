```python
from sklearn.datasets import load_digits
```

```python
# The perceptron is a single processing unit of any neural network. Frank Rosenblatt first proposed in 1958 
# is a simple neuron which is used to classify its input into one or two categories. 
# Perceptron is a linear classifier, and is used in supervised learning. It helps to organize the given input data.

# The weights are initialized with the random values at the origination of each training.
# For each element of the training set, the error is calculated with the difference between the desired output and the actual output. 
# The calculated error is used to adjust the weight.
# The process is repeated until the fault made on the entire training set is less than the specified limit until the maximum 
# number of iterations has been reached

# Perceptron is a classification algorithm which shares the same underlying implementation with SGDClassifier
# The Perceptron is another simple classification algorithm suitable for large scale learning. By default:

# It does not require a learning rate.

# It is not regularized (penalized).

# It updates its model only on mistakes.

# The last characteristic implies that the Perceptron is slightly faster to train than SGD with the hinge loss
# and that the resulting models are sparser.
```

```python
digits = load_digits()  # mnist data(pixel range 0 to 255) 
```

```python
digits
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
X = digits.data
y = digits.target
```

```python
# machine check the image by pixel value
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = .80, random_state =1)
```

```python
from sklearn.linear_model import Perceptron
model = Perceptron()
```

```python
from sklearn.metrics import classification_report
```

```python
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
```

```python
print(accuracy_score(y_test, predicted))
```

```python
print(model.score(X_test, y_test))
# Note model.score will give accuracy score between Y_test & predicted value. 
# However, we need to specify parameters as X_test and y _test
# see above block and this accuracy_score is same with both methods
```

```python
print(model.score(X_train, y_train))
```

```python
# Confusion Matrix
print(confusion_matrix(y_test, predicted))
```

```python
# Classification Report
print(classification_report(y_test, predicted))
```

# Stochastic Gradient Descent (SGD)

```python
# Stochastic Gradient Descent (SGD) is a simple yet very efficient approach to fitting linear classifiers
 # this is will give similar result as perceptron using "perceptron as loss function"
# Stochastic Gradient Descent is sensitive to feature scaling, so it is highly recommended to scale your data.
# This can be easily done using StandardScaler:
from sklearn.linear_model import SGDClassifier
sdg_p = SGDClassifier(loss="perceptron", eta0=1, learning_rate="constant", penalty=None)
```

```python
sdg_p.fit(X_train, y_train)
```

```python
predicted_sdg = sdg_p.predict(X_test)
```

```python
print(confusion_matrix(y_test, predicted_sdg))
```

```python
print(classification_report(y_test, predicted_sdg))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
classifier.fit(X_train, y_train)  
```

```python
y_pred_test_knn = classifier.predict(X_test)  
y_pred_train_knn = classifier.predict(X_train)  
```

```python
print(classification_report(y_test, y_pred_test_knn))
```

```python
print(classification_report(y_train, y_pred_train_knn))
```

```python
# Confusion Matrix
print(confusion_matrix(y_test, y_pred_test_knn))
```
